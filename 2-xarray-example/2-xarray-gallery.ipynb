## 读取多个 nc 文件

gallery-data文件夹下面 14303200 和 14306340 下面多个nc文件，读取到一个xarray-dataset中，这两个文件夹的数据分别属于 14303200 和 14306340 这两个对象。

xarray读取多文件的函数open_mfdataset 需要同时指定多文件，所以为了读取不同的文件夹下的文件，需要首先一次性将符合条件的数据读取出来。

或者同一个文件夹读取多个文件，然后再将多个dataset连接起来（这种更灵活一些）。

主要参考以下资料：

- [how to open multiple netcdf files stored in multiple folders Python](https://stackoverflow.com/questions/50096170/how-to-open-multiple-netcdf-files-stored-in-multiple-folders-python)
- [python-xarray: open_mfdataset concat along two dimensions](https://stackoverflow.com/questions/47545138/python-xarray-open-mfdataset-concat-along-two-dimensions)

可以看到因为x和y的坐标在两个文件夹下的nc文件里是不同的，所以当一次性读取多文件的时候，是会把两个文件的x和y合起来的，这就会造成有很多NaN值的出现，因为每个nc文件都只在自己对应的x和y坐标下才有值。

这是需要**特别注意**的！所以对于文件内容坐标有很多不同的情况，要慎重使用 open_mfdataset 函数。

这里给出的例子重点是为了记录使用open_mfdataset函数读取多文件数据，所以下面就暂时不再多考虑这一情况了。

上面的方式比较适合同一类型的在不同文件夹下的文件，如果不同文件夹下内容有些不同，需要一个维度标注出来，还是下面的方式更方便一些。

```python
import os
import pandas as pd
import xarray as xr

dir_list = ["14303200","14306340"]
ens_list = []
for num in range(len(dir_list)):
    ens = dir_list[num]
    ens_list.append(xr.open_mfdataset(os.path.join("gallery-data", ens, '*.nc')))
new_dim = pd.Index(dir_list)
# dim的值如果是新维度，会增加到第一个维度上
ds = xr.concat(ens_list, dim=new_dim)
```

```python
ens_list
```

```python
ds
```

看看合并之后的数据是不是仍能顺利地索引到正确的数据。

比如只取出第一个文件夹下读出的内容，即 concat_dim 值为"14303200"的。

```python
ds.sel(concat_dim="14303200")
```

如前面所述，concatenate后x和y是有所变化的，所以直接判断肯定是不相等。不过当做减法时，是在对应坐标上做运算，所以从具体数据上看，是一致的：

```python
(ds.sel(concat_dim="14303200").swe-(ens_list[0].swe)).values
```

可以看到做减法是在x和y对应坐标上做的：

```python
(ds.sel(concat_dim="14303200").swe-(ens_list[0].swe)).values.shape
```

## xarray中的索引

根据[文档介绍](https://xarray.pydata.org/en/stable/user-guide/indexing.html#indexing)，xarray中dataset和dataarray的索引形式有几种：

|Dimension lookup|Index lookup|DataArray syntax|Dataset syntax|
|-|-|-|-|
|Positional|By integer|da[:, 0]|not available|
|Positional|By label|da.loc[:, 'IA']|not available|
|By name|By integer|da.isel(space=0) or da[dict(space=0)]|ds.isel(space=0) or ds[dict(space=0)]|
|By name|By label|da.sel(space='IA') or da.loc[dict(space='IA')]|ds.sel(space='IA') or ds.loc[dict(space='IA')]|

简单归纳下，Dataset只能用name来进行dimension索引，DataArray适合各种针对 dimension 和index的索引。

比如下面要索引日期，注意是闭区间：

```python
ds_1990.sel(time=slice('1990-03-01', '1990-03-05'))
```

## 插值

上面用到的实例数据来自daymet，daymet数据在闰年也只有365天数据，12月31日那天是直接被忽略掉的，为了保持使用时的完整性和便利性，考虑在闰年的最后一天插值该时间序列数据。

下面先从简单的例子开始。主要参考了：[xarray User Guide / Interpolating data](https://xarray.pydata.org/en/stable/user-guide/interpolation.html)

```python
import numpy as np
import xarray as xr
```

```python
da = xr.DataArray(
    np.sin(0.3 * np.arange(12).reshape(4, 3)),
    [("time", np.arange(4)), ("space", [0.1, 0.2, 0.3])],
)
da
```

插值的方式和索引的方式是类似的，下面是索引

```python
da.sel(time=3)
```

现在在第2和3之间时段之间插值，也就是第三行和第四行之间：

```python
da.interp(time=2.5)
```

也可以像索引那样接受一个array-like的参数

```python
da.sel(time=[2, 3])
```

```python
da.interp(time=[2.5, 3.5])
```

如果是沿着numpy.datetime64的坐标插值，可以传字符串：

```python
import pandas as pd
da_dt64 = xr.DataArray(
    [1, 3], [("time", pd.date_range("1/1/2000", "1/3/2000", periods=2))]
)
da_dt64
```

```python
da_dt64.interp(time="2000-01-02")
```

通过指定需要的时间段，可以将新插入的值合并入原始 DataArray 

```python
da_dt64.interp(time=pd.date_range("1/1/2000", "1/3/2000", periods=3))
```

如果是在时间范围外插值，则默认插入的是NaN

```python
da_dt64.interp(time="2000-01-04")
```

如果想要外延插值，则可以进一步利用interpolate_na

```python
da_dt64_new = da_dt64.interp(time=pd.date_range("1/1/2000", "1/4/2000", periods=4))
da_dt64_new
```

```python
da_dt64_new.interpolate_na(dim="time",fill_value="extrapolate")
```

interpolate_na 同样不是 in-place的操作，如果需要更新其值，需要赋值

```python
da_dt64_new
```

```python
da_dt64_new_nonan = da_dt64_new.interpolate_na(dim="time",fill_value="extrapolate")
da_dt64_new_nonan
```

下面是多维数据的情况：

```python
da.interp(time=2.5, space=0.15)
```

```python
da.interp(time=[1.5, 2.5], space=[0.15, 0.25])
```

xarray这些插值函数具体使用的插值方法是 scipy.interpolate.interp1d （一维数据）和 scipy.interpolate.interpn() （多维数据）

下面看看本文实例数据。补充1996年12月31日的数据。这里选择14303200文件夹下数据为例。

```python
import glob
import os

years = ["1996","1997"]
ens_19961997 = "14303200"
name_lst=[]
for name in glob.glob(os.path.join("gallery-data", ens_19961997,'*.nc')):
    if name.split("_")[1] in years:
        name_lst.append(name)
    # 只要文件名称对应就行了    
ds_19961997 = xr.open_mfdataset(name_lst)
ds_19961997
```

可以看到731天只有730天的数据，1996年12月31日并没有数据，比如取出dayl看一看

```python
ds_19961997.sel(time="1996-12-31")["dayl"].values
```

下面尝试通过指定时间段，插入值并合并入原始 Dataset。因为interp不是inp-place操作，所以需要返回到一个新的dataset中。注意这里是闭区间，periods是pandas date_range的一部分。

另外需要注意原始数据的时间是12点的，所以时间也必须得是12点，否则插值的时候每个时段默认是0点，全部都会插值的，并且由于第一个时段00点在现在的第一个时段12点之前，会插值NaN。

这点要特别注意，所以最好的办法还是将时间构造到12点（也可以使用原数据中的时间，但是比如如果正好最后一个时段在原数据时段外，那么就这样就不行了，不过这里处于方便，暂时就直接这么用了）。

```python
ds_19961997["time"].values[0]
```

```python
ds_new = ds_19961997.interp(time=pd.date_range(ds_19961997["time"].values[0], ds_19961997["time"].values[-1], periods=731))
```

```python
ds_new
```

```python
ds_new.sel(time="1996-12-31")["dayl"].values
```

检查一下原来的值是否保持：

```python
ds_new.sel(time="1996-01-01")["dayl"].values == ds_19961997.sel(time="1996-01-01")["dayl"].values
```

## xarray与pandas和numpy的转换

xarray的数据是懒加载地，通常我们还是需要把数据取出来用，有时候需要将数据转换为pandas DataFrame或者numpy array或者torch的tensor等。

比如想要取出 dayl, prcp 的值，并组成一个numpy array

```python
import os
import glob
import pandas as pd
import xarray as xr
years = ["1990","1991"]
dir_list = ["14303200","14306340"]
ens_list_19901991 = []
for num in range(len(dir_list)):
    ens_19901991 = dir_list[num]
    name_lst=[]
    for name in glob.glob(os.path.join("gallery-data", ens_19901991,'*.nc')):
        if name.split("_")[1] in years:
            name_lst.append(name)
    # 只要文件名称对应就行了    
    ens_list_19901991.append(xr.open_mfdataset(name_lst))
# dim的值如果是新维度，会增加到第一个维度上
new_dim = pd.Index(dir_list)
ds_19901991 = xr.concat(ens_list_19901991, dim=new_dim)
```

```python
ds_19901991[['dayl', 'prcp']]
```

## DataArray基本运算

对 xarray 中 DataArray做计算时, 默认的是按照维度名字匹配的, 这里以乘法为例，看下面的例子.

```python
import xarray as xr
x = xr.DataArray(
     [[1.0, 2.0], [3.0, 5.0]],
     dims=("lat", "lon"),
     coords={"lat": [35.0, 40.0], "lon": [100.0, 120.0]},
     name="var1",
)
x
```

```python
t = xr.DataArray([0.1,0.2,0.3], dims="time", coords={"time": [1, 2, 3]}, name="var2")
t
```

```python
y = x * t
y
```

可以看到上述计算中, 乘积是执行的笛卡尔积, 因为没有维度对应上. 再看一个例子

```python
z = x * y
z
```

```python
x
```

可以看到, lat和lon一致的地方对应相乘; 关于时间维, 相当于在x上执行了广播运算. 也就是说相当于 x * y.isel(time=0),  x * y.isel(time=1),  x * y.isel(time=2)

```python
x * y.isel(time=0)
```
