## 读取多个 nc 文件

gallery-data文件夹下面 14303200 和 14306340 下面共4个nc文件，读取到一个xarray-dataset中，这两个文件夹的数据分别属于 14303200 和 14306340 这两个对象。

xarray读取多文件的函数open_mfdataset 需要同时指定多文件，所以为了读取不同的文件夹下的文件，需要首先一次性将符合条件的数据读取出来。

或者同一个文件夹读取多个文件，然后再将多个dataset连接起来（这种更灵活一些）。

主要参考以下资料：

- [how to open multiple netcdf files stored in multiple folders Python](https://stackoverflow.com/questions/50096170/how-to-open-multiple-netcdf-files-stored-in-multiple-folders-python)
- [python-xarray: open_mfdataset concat along two dimensions](https://stackoverflow.com/questions/47545138/python-xarray-open-mfdataset-concat-along-two-dimensions)

```python
import glob

# Get a list of all .nc files available in different folders
filenames = glob.glob("gallery-data/*/*.nc")
filenames
```

```python
import xarray as xr
dsmerged = xr.open_mfdataset(filenames)
```

可以看到因为x和y的坐标在两个文件夹下的nc文件里是不同的，所以当一次性读取多文件的时候，是会把两个文件的x和y合起来的，这就会造成有很多NaN值的出现，因为每个nc文件都只在自己对应的x和y坐标下才有值。

这是需要**特别注意**的！所以对于文件内容坐标有很多不同的情况，要慎重使用 open_mfdataset 函数。

这里给出的例子重点是为了记录使用open_mfdataset函数读取多文件数据，所以下面就暂时不再多考虑这一情况了。

上面的方式比较适合同一类型的在不同文件夹下的文件，如果不同文件夹下内容有些不同，需要一个维度标注出来，还是下面的方式更方便一些。

```python
import os
import pandas as pd
import xarray as xr

dir_list = ["14303200","14306340"]
ens_list = []
for num in range(len(dir_list)):
    ens = dir_list[num]
    ens_list.append(xr.open_mfdataset(os.path.join("gallery-data", ens, '*.nc')))
new_dim = pd.Index(dir_list)
# dim的值如果是新维度，会增加到第一个维度上
ds = xr.concat(ens_list, dim=new_dim)
```

```python
ens_list
```

```python
ds
```

看看合并之后的数据是不是仍能顺利地索引到正确的数据。

比如只取出第一个文件夹下读出的内容，即 concat_dim 值为"14303200"的。

```python
ds.sel(concat_dim="14303200")
```

如前面所述，concatenate后x和y是有所变化的，所以直接判断肯定是不相等。不过当做减法时，是在对应坐标上做运算，所以从具体数据上看，是一致的：

```python
(ds.sel(concat_dim="14303200").swe-(ens_list[0].swe)).values
```

可以看到做减法是在x和y对应坐标上做的：

```python
(ds.sel(concat_dim="14303200").swe-(ens_list[0].swe)).values.shape
```

稍微复杂一点，比如只选择两个文件夹1990年的那些nc文件，则可以利用文件名索引：

```python
ens_list_1990 = []
for num in range(len(dir_list)):
    ens_1990 = dir_list[num]
    # 只要文件名称对应就行了    
    ens_list_1990.append(xr.open_mfdataset(os.path.join("gallery-data", ens_1990, '*_1990_*.nc')))
# dim的值如果是新维度，会增加到第一个维度上
ds_1990 = xr.concat(ens_list_1990, dim=new_dim)
```

```python
ds_1990
```

如果需要具体到内部的日期，可以采用索引的方式（参考了[这里](https://stackoverflow.com/questions/51976126/subset-an-xarray-dataset-or-data-array-in-the-time-dimension)），注意是闭区间：

```python
ds_1990.sel(time=slice('1990-03-01', '1990-03-05'))
```

再比如如果有多年数据，只允许选择1990和1991的，可以这样：

```python
years = ["1990","1991"]
ens_list_19901991 = []
for num in range(len(dir_list)):
    ens_19901991 = dir_list[num]
    name_lst=[]
    for name in glob.glob(os.path.join("gallery-data", ens_19901991,'*.nc')):
        if name.split("_")[1] in years:
            name_lst.append(name)
    # 只要文件名称对应就行了    
    ens_list_19901991.append(xr.open_mfdataset(name_lst))
# dim的值如果是新维度，会增加到第一个维度上
ds_19901991 = xr.concat(ens_list_19901991, dim=new_dim)
```

```python
ds_19901991
```

## xarray与pandas和numpy的转换

xarray的数据是懒加载地，通常我们还是需要把数据取出来用，有时候需要将数据转换为pandas DataFrame或者numpy array或者torch的tensor等。

比如想要取出 dayl, prcp 的值，并组成一个numpy array

```python
ds_19901991[['dayl', 'prcp']]
```

dataset可以转成dataarray，此数据集的数据变量将相互广播并沿新数组的第一个轴堆叠。此数据集的所有坐标将保持。

```python
ds_trans_origin = ens_list[0]
```

```python
ds_dataarray = ds_trans_origin[['dayl', 'prcp']].to_array()
```

```python
ds_dataarray
```

dataarray可以转成numpy的array，直接取values即可。可以看到维度是保持顺序的。

```python
ds_dataarray.values.shape
```

当选择的变量对应的数据的维度一致时，可以直接转为dataframe

```python
ds_trans_origin[['dayl', 'prcp']].to_dataframe()
```

## DataArray基本运算

对 xarray 中 DataArray做计算时, 默认的是按照维度名字匹配的, 这里以乘法为例，看下面的例子.

```python
import xarray as xr
x = xr.DataArray(
     [[1.0, 2.0], [3.0, 5.0]],
     dims=("lat", "lon"),
     coords={"lat": [35.0, 40.0], "lon": [100.0, 120.0]},
     name="var1",
)
x
```

```python
t = xr.DataArray([0.1,0.2,0.3], dims="time", coords={"time": [1, 2, 3]}, name="var2")
t
```

```python
y = x * t
y
```

可以看到上述计算中, 乘积是执行的笛卡尔积, 因为没有维度对应上. 再看一个例子

```python
z = x * y
z
```

```python
x
```

可以看到, lat和lon一致的地方对应相乘; 关于时间维, 相当于在x上执行了广播运算. 也就是说相当于 x * y.isel(time=0),  x * y.isel(time=1),  x * y.isel(time=2)

```python
x * y.isel(time=0)
```
